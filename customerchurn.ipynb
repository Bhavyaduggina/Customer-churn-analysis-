```python
car=df.select_dtypes(include='object').columns.tolist()
for var in car:
  print(df[var].value_counts())
```

```python
numcols=len(car)
numrows=(numcols+2)//3
fig,axs=plt.subplots(nrows=numrows,ncols=3,figsize=(15,5*numrows))
axs=axs.flatten()
for i,var in enumerate(car):
    top_values = df[var].value_counts().index
    filtered_df = df[df[var].isin(top_values)]
    sns.countplot(x=var, data=filtered_df, ax=axs[i])
    axs[i].set_title(var)
    axs[i].tick_params(axis='x', rotation=90)
if numcols < len(axs):
    for i in range(numcols, len(axs)):
        fig.delaxes(axs[i])
fig.tight_layout()
plt.show()
```

```python
x = df.drop('Churn Value', axis=1)
y = df['Churn Value']
from sklearn.preprocessing import RobustScaler
r=RobustScaler()
xscale=pd.DataFrame(r.fit_transform(x),columns=x.columns)
```

```python
xscale.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(xscale,y,test_size=0.3,random_state=0)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state = 2)
xtrain, ytrain= sm.fit_resample(xtrain, ytrain)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
from scipy import stats
selected_columns = ['Avg Monthly GB Download', 'Total Refunds', 'Total Extra Data Charges', 'Total Long Distance Charges']
z_scores = np.abs(stats.zscore(xtrain[selected_columns]))
threshold = 3
outlier_indices = np.where(z_scores > threshold)[0]
xtrain = xtrain.drop(xtrain.index[outlier_indices])
ytrain = ytrain.drop(ytrain.index[outlier_indices])
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(class_weight='balanced')
param_grid = {
    'max_depth': [3, 4, 5, 6, 7, 8],
    'min_samples_split': [2, 3, 4],
    'min_samples_leaf': [1, 2, 3, 4],
    'random_state': [0, 42]
}
grid=GridSearchCV(dtree,param_grid,cv=5)
grid.fit(xtrain,ytrain)

print(grid.best_params_)
```

```python
tree=DecisionTreeClassifier(max_depth=8, min_samples_leaf= 4, min_samples_split=2,random_state=42)
tree.fit(xtrain,ytrain)
ypred=tree.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(ytest,ypred)
```

```python
print(classification_report(ytest,ypred))
```

```python
pd.crosstab(ytest,ypred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(class_weight='balanced')
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 5, 10],
    'max_features': ['sqrt', 'log2', None],
    'random_state': [0, 42]
}
grid_search = GridSearchCV(rfc, param_grid, cv=5)
grid_search.fit(xtrain, ytrain)
print(grid_search.best_params_)
```

```python
rf=RandomForestClassifier(max_depth=None, max_features= 'sqrt',n_estimators= 200, random_state=0)
rf.fit(xtrain,ytrain)
ypred2=rf.predict(xtest)
```

```python
accuracy_score(ytest,ypred2)
```

```python
print(classification_report(ytest,ypred2))
```

```python
pd.crosstab(ytest,ypred2)
```
